```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
X = df[['area']]
y = df['price']
```

```python
model = DecisionTreeRegressor()
model.fit(X, y)
```

```python
pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, pred)
plt.xlabel('area')
plt.ylabel('price')
plt.title('Decision Tree Regression')
plt.show()
```
